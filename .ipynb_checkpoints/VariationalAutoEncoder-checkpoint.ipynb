```python
import numpy as np
from tensorflow.keras.datasets import fashion_mnist
from scipy.ndimage import gaussian_filter, sobel, median_filter, uniform_filter

# Cargar y preparar datos
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

# Función para aplicar una operación canal por canal
def apply_filter_batch(x, func, **kwargs):
    return np.stack([func(x[i, ..., 0], **kwargs) for i in range(x.shape[0])], axis=0)[..., np.newaxis]

# Generar 11 canales distintos
def generate_extra_channels(x):
    channels = [x]
    channels.append(apply_filter_batch(x, gaussian_filter, sigma=0.6))
    channels.append(apply_filter_batch(x, gaussian_filter, sigma=1.2))
    channels.append(apply_filter_batch(x, sobel, axis=0))
    channels.append(apply_filter_batch(x, sobel, axis=1))
    channels.append(apply_filter_batch(x, median_filter, size=3))
    channels.append(apply_filter_batch(x, median_filter, size=5))
    channels.append(apply_filter_batch(x, uniform_filter, size=3))
    channels.append(apply_filter_batch(x, uniform_filter, size=5))
    channels.append(apply_filter_batch(x, lambda img: img**2))
    channels.append(apply_filter_batch(x, lambda img: np.clip(img + 0.1 * np.random.randn(*img.shape), 0, 1)))
    channels.append(apply_filter_batch(x, lambda img: np.flip(img, axis=1)))  # Flip horizontal

    return np.concatenate(channels, axis=-1)

# Aplicar al dataset
x_train_12ch = generate_extra_channels(x_train)
x_test_12ch = generate_extra_channels(x_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_reconstruction_12ch(model, x, index=0):
    """
    model : VAE entrenado
    x     : datos de entrada (N, H, W, 12)
    index : índice de la muestra a visualizar
    """

    # Seleccionamos una muestra
    x_real = x[index:index+1]

    # Reconstrucción
    x_rec = model.predict(x_real, verbose=0)

    fig, axes = plt.subplots(2, 12, figsize=(24, 4))

    for ch in range(12):
        # Fila 1: imagen real
        axes[0, ch].imshow(x_real[0, :, :, ch], cmap="gray")
        axes[0, ch].set_title(f"Real ch {ch}")
        axes[0, ch].axis("off")

        # Fila 2: reconstrucción
        axes[1, ch].imshow(x_rec[0, :, :, ch], cmap="gray")
        axes[1, ch].set_title(f"Recon ch {ch}")
        axes[1, ch].axis("off")

    axes[0, 0].set_ylabel("Real", fontsize=12)
    axes[1, 0].set_ylabel("Recon", fontsize=12)

    plt.tight_layout()
    plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

# -------------------------------------------------
# Sampling layer (reparametrization trick)
# -------------------------------------------------
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        epsilon = tf.random.normal(shape=tf.shape(z_mean))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon


# -------------------------------------------------
# Variational Convolutional Autoencoder
# -------------------------------------------------
def modelo_VCAE(input_shape=(28, 28, 12), latent_dim=16, lr=1e-3):
    conv_size = 32
    dropout_val = 0.25

    # ---------------- Encoder ----------------
    encoder_inputs = layers.Input(shape=input_shape, name="encoder_input")

    x = layers.Conv2D(conv_size, 3, padding="same", activation="relu")(encoder_inputs)
    x = layers.Dropout(dropout_val)(x)
    x = layers.Conv2D(conv_size, 3, padding="same", activation="relu")(x)
    x = layers.Dropout(dropout_val)(x)
    x = layers.Flatten()(x)

    z_mean = layers.Dense(latent_dim, name="z_mean")(x)
    z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
    z = Sampling(name="z")([z_mean, z_log_var])

    encoder = Model(
        encoder_inputs, [z_mean, z_log_var, z], name="encoder"
    )

    # ---------------- Decoder ----------------
    latent_inputs = layers.Input(shape=(latent_dim,), name="decoder_input")

    x = layers.Dense(
        conv_size * input_shape[0] * input_shape[1],
        activation="relu"
    )(latent_inputs)
    x = layers.Reshape((input_shape[0], input_shape[1], conv_size))(x)
    x = layers.Conv2DTranspose(conv_size, 3, padding="same", activation="relu")(x)
    x = layers.Dropout(dropout_val)(x)
    x = layers.Conv2DTranspose(conv_size, 3, padding="same", activation="relu")(x)
    decoder_outputs = layers.Conv2DTranspose(
        input_shape[-1],
        3,
        padding="same",
        activation="sigmoid",
        name="decoder_output"
    )(x)

    decoder = Model(latent_inputs, decoder_outputs, name="decoder")

    # ---------------- VAE Model ----------------
    class VAE(tf.keras.Model):
        def __init__(self, encoder, decoder, **kwargs):
            super().__init__(**kwargs)
            self.encoder = encoder
            self.decoder = decoder

            self.total_loss_tracker = tf.keras.metrics.Mean(name="total_loss")
            self.reconstruction_loss_tracker = tf.keras.metrics.Mean(
                name="reconstruction_loss"
            )
            self.kl_loss_tracker = tf.keras.metrics.Mean(name="kl_loss")

        @property
        def metrics(self):
            return [
                self.total_loss_tracker,
                self.reconstruction_loss_tracker,
                self.kl_loss_tracker,
            ]

        def call(self, inputs, training=False):
            z_mean, z_log_var, z = self.encoder(inputs, training=training)
            return self.decoder(z, training=training)

        def _rmse(self, x, x_hat):
            # RMSE por muestra
            mse = tf.reduce_mean(
                tf.square(x - x_hat),
                axis=(1, 2, 3)
            )
            return tf.sqrt(mse + 1e-8)

        def train_step(self, data):
            if isinstance(data, tuple):
                data = data[0]

            with tf.GradientTape() as tape:
                z_mean, z_log_var, z = self.encoder(data, training=True)
                reconstruction = self.decoder(z, training=True)

                recon_loss = tf.reduce_mean(
                    self._rmse(data, reconstruction)
                )

                kl_loss = -0.5 * tf.reduce_mean(
                    tf.reduce_sum(
                        1 + z_log_var
                        - tf.square(z_mean)
                        - tf.exp(z_log_var),
                        axis=1,
                    )
                )

                total_loss = recon_loss + kl_loss

            grads = tape.gradient(total_loss, self.trainable_weights)
            self.optimizer.apply_gradients(
                zip(grads, self.trainable_weights)
            )

            self.total_loss_tracker.update_state(total_loss)
            self.reconstruction_loss_tracker.update_state(recon_loss)
            self.kl_loss_tracker.update_state(kl_loss)

            return {
                "loss": self.total_loss_tracker.result(),
                "reconstruction_loss": self.reconstruction_loss_tracker.result(),
                "kl_loss": self.kl_loss_tracker.result(),
            }

        def test_step(self, data):
            if isinstance(data, tuple):
                data = data[0]

            z_mean, z_log_var, z = self.encoder(data, training=False)
            reconstruction = self.decoder(z, training=False)

            recon_loss = tf.reduce_mean(
                self._rmse(data, reconstruction)
            )

            kl_loss = -0.5 * tf.reduce_mean(
                tf.reduce_sum(
                    1 + z_log_var
                    - tf.square(z_mean)
                    - tf.exp(z_log_var),
                    axis=1,
                )
            )

            total_loss = recon_loss + 0.1*kl_loss

            self.total_loss_tracker.update_state(total_loss)
            self.reconstruction_loss_tracker.update_state(recon_loss)
            self.kl_loss_tracker.update_state(kl_loss)

            return {
                "loss": self.total_loss_tracker.result(),
                "reconstruction_loss": self.reconstruction_loss_tracker.result(),
                "kl_loss": self.kl_loss_tracker.result(),
            }

    vae = VAE(encoder, decoder, name="VCAE")
    vae.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=lr)
    )

    return vae
```

```python
x_train_12ch = x_train_12ch.astype("float32")
x_train_12ch = (x_train_12ch - x_train_12ch.min()) / (
    x_train_12ch.max() - x_train_12ch.min()
)

```

```python
N = x_train_12ch.shape[0]
idx = np.random.choice(N, size=int(0.1 * N), replace=False)

x_train_12ch_sub = x_train_12ch[idx]
```

```python
vae = modelo_VCAE(
    input_shape=(28, 28, 12),
    latent_dim=4,
    lr=1e-3
)

history = vae.fit(
    x_train_12ch_sub,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    shuffle=True
)


```

```python
plot_reconstruction_12ch(vae, x_train_12ch_sub, index=0)
```
